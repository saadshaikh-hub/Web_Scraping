```python
# notebooks/data_analysis.ipynb

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data
data_path = '../data/cleaned_data.csv'
data = pd.read_csv(data_path)

# Display the first few rows of the dataset
print("Cleaned Data Preview:")
data.head()

# Set Seaborn style
sns.set(style="whitegrid")

# 1. Bar Plot for Book Prices
plt.figure(figsize=(12, 6))
# Convert price to float for plotting
data['Price'] = data['Price'].replace({'£': '', '0': ''}, regex=True).astype(float)
sns.barplot(x='Book Title', y='Price', data=data.sort_values('Price', ascending=False).head(10))
plt.title('Top 10 Most Expensive Books')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Book Title')
plt.ylabel('Price (£)')
plt.tight_layout()
plt.show()

# 2. Count Plot for Ratings
plt.figure(figsize=(8, 5))
sns.countplot(x='Rating', data=data)
plt.title('Distribution of Book Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# 3. Box Plot for Prices by Rating
plt.figure(figsize=(12, 6))
sns.boxplot(x='Rating', y='Price', data=data)
plt.title('Price Distribution by Rating')
plt.xlabel('Rating')
plt.ylabel('Price (£)')
plt.show()
```
